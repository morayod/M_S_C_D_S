### Import libraries

```python
from nltk.corpus import stopwords
import pandas as pd
from arabica import arabica_freq
from arabica import cappuccino
import seaborn as sns
```

### Load data

```python
url = 'https://raw.githubusercontent.com/morayod/M_S_C_D_S/main/reviews.csv'
df = pd.read_csv(url)
df.head(10)
```

### Create unigrams, bigrams and trigrams

```python
result = arabica_freq(text = df['Review'],
             time = df['Time_submitted'],
             date_format = 'us',
             time_freq = 'ungroup',
             max_words = 500,
             stopwords = ['english'],
             skip = ['<br />'],
             numbers = True,
             lower_case = True)
```

```python
result
```

### Word cloud with bigrams 

```python
cappuccino(text = df['Review'],
           time = df['Time_submitted'],
           date_format = 'us',
           plot = 'wordcloud',
           ngram = 2,
           time_freq = 'ungroup',
           max_words = 150,
           stopwords = ['english'],
           skip = ['music','song','app','songs','spotify','play','listen'],
           numbers = True,
           lower_case = True)
```
